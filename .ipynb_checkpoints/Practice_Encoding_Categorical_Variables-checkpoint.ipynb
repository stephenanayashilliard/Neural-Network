# Practice Encoding Categorical Variables

```python
# Import our dependencies
import pandas as pd
import sklearn as skl
```

```python
# Read in our ramen data
ramen_df = pd.read_csv("ramen-ratings.csv")
ramen_df.head()
```

```python
# Print out the Country value counts
country_counts = ramen_df.Country.value_counts()
country_counts
```

Looking at the country's unique value counts, there are a number of countries that appear frequently in the dataset such as Japan, USA, and South Korea. However, many countries appear semi-frequently and even more rarely appear in the dataset.

How do we determine which countries are uncommon enough to bucket into the "other" category? The most straightforward method is to use a density plot to identify where the value counts "fall off" and set the threshold within this region

```python
# Visualize the value counts by producing density plot.
country_counts.plot.density()
```

According to the density plot, the most common unique values have more than 100 instances within the dataset. Therefore, we can bucket any country that appears fewer than 100 times in the dataset as "other." To do this, we'll use a Python for loop and Pandas' replace method. 

```python
# Determine which values to replace
replace_countries = list(country_counts[country_counts < 100].index)
```

```python
# Replace in DataFrame
for country in replace_countries:
    ramen_df.Country = ramen_df.Country.replace(country,"Other")
```

```python
# Check to make sure binning was successful
ramen_df.Country.value_counts()
```

## OneHotEncoder
Now that we have reduced the number of unique values in the country variable, we're ready to transpose the variable using one-hot encoding. The easiest way to perform one-hot encoding in Python is to use Scikit-learn's OneHotEncoder module on the country variable. To build the encoded columns, we must create an instance of OneHotEncoder and "fit" the encoder with our values. Again, we must add and run the following code:

```python
# Create the OneHotEncoder instance
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(sparse=False)
```

```python
# Fit the encoder and produce encoded DataFrame
encode_df = pd.DataFrame(enc.fit_transform(ramen_df.Country.values.reshape(-1,1)))
```

```python
# Rename encoded columns
encode_df.columns = enc.get_feature_names(['Country'])
encode_df.head()
```

## Merge the DataFrames
We must join the encoded DataFrame with the original and drop the original "Country" column. The process of joining the two DataFrames together is handled by the Pandas merge method and can be performed within one line.

```python
# Merge the two DataFrames together and drop the Country column
ramen_df.merge(encode_df,left_index=True,right_index=True).drop("Country",1)
```
